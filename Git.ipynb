# Git
## Main commands

#### настройка
    git config --global user.name "Name"
    git config --global user.email email@example.com
    
#### создание нового каталога, создание репозитория
    mkdir Desktop/git_example/
    cd Desktop/git_example/
    git init
    
#### подготовка файлов и фиксация изменений
    git add [file_name.file_extension] --- добавить файлы рабочей директории в индекс
        git add *.[file_extension] --- добавить все файлы с указанным расширением в индекс
        git add ./[folder] --- добавить указанный каталог в индекс
        git add -A --- добавить все файлы в директории
        
        создание файла .gitignore --- игнорировать отдельные файлы и каталоги при добавлении:
                    *.log   --- игнорировать файлы с расширением .log
                    build/  --- игнорировать файлы в указанном каталоге
            git add .gitignore
            git commit -m “added .gitignore file”
    git rm [file_name.file_extension]… --- удалить указанные файлы из индекса
    
    
    git commit -m “comment” --- создать коммит (в текущей ветке) с комментарием
        git stash --- сохранить незакомиченные изменения и сбросить рабочую копию до HEAD'а репозитория (чтобы можно было переключиться на другую ветку без создания коммита)
    
    git reset HEAD^ --- отмена коммита(-ов) в локальном репозитории
    git revert HEAD --- создание противоположного текущему коммита для удалённого репозитория (для внесения изменений)
        git revert [commit]
    
#### ветвление
    git checkout [branch_name] --- переключение на ветку (и выгрузка содержимого в рабочую директорию)
    git branch [branch_name] --- создание ветки на основе текущей
        git checkout -b [branch_name] --- создать новую ветку и переключиться на неё
        git checkout -b [branch_name] [commit] --- создать новую ветку с определённого коммита и переключиться на неё
    git branch -a --- список веток в репозитеории с указанием текущей
    git branch -D [branch_name] --- удалить ветку
    
    git merge [branch_name] --- смерджить ветку в текущую (и переключиться на результирующий коммит)
    git rebase [destination_branch_name] [branch_name] --- скопировать ветку в другую ветку
        git rebase [destination_branch_name] --- скопировать текущую ветку в указанную

    git cherry-pick [commit]… --- копировать коммиты в HEAD (только если коммит не является предком HEAD)

    git rebase -i HEAD~[num] --- изменить (переставить/выкинуть/слить в один) последние num коммитов
            внесение изменений в старый коммит с помощью rebase -i :
            1. переставить коммит так, чтобы нужный находился наверху при помощи git rebase -i
            2. внести изменения при помощи git commit --amend
            3. переставить всё обратно при помощи git rebase -i
    
#### перемещение по дереву коммитов
    git checkout [commit] --- отделить HEAD от конца ветки к определённому коммиту (достаточно указать первые несколько символов коммита (его хэша))
    git checkout [commit]^ --- переместиться на один коммит назад
        git checkout HEAD^
        git checkout [commit]^^ --- переместиться на прародителя
        git checkout main^2 --- переместиться на второго родителя (в мерджевом коммите) (по умолчанию - на первого родителя)
                git checkout HEAD~^2~2 --- пример совместного использования команд
    git checkout [commit]~[num] --- переместиться на num коммитов назад
    git branch -f [branch_name] HEAD~3 --- branch forcing: переместить (принудительно) ветку (указатель на ветку) на три родителя назад от HEAD; -f прикрепляет ветку к коммиту

#### работа с удалённым репозиторием
    git clone --- создать локальную копию удалённого репозитория
        git clone ; git checkout -b <имя ветки> origin/<имя ветки> --- когда нужна определённая ветка, а не всё хранилище
    git remote add origin [URL] --- присвоить сокращение адресу удалённого репозитория
    
    git fetch --- извлечь данные из удалённого репозитория (синхронизация, но без слияния со своими ветками)
        git fetch origin [branch_name] --- скопировать отсутствующие коммиты из ветки branch_name удалённого репозитория в ветку origin/[branch_name] локального репозитория
            git fetch [<источник>]:[<получатель>] --- (принцип аналогичен команде git push origin [<источник>]:[<получатель>] (см.ниже)) (для выполнения нельзя находится на ветке-получателе) (при отсутствии источника создаётся ветка)
    git pull --- fetch + merge
            git pull origin [<источник>]:[<получатель>] --- (аналогично fetch) + merge
    git push --- противоположность pull (без аргументов варьируется в зависимости от значения push.default, указанной в настройках git)
        git push origin [branch_name] --- копировать отсутствующие в соответствующей ветке удалённого репозитория коммиты из ветки branch_name локального репозитория
            git push origin [<источник>]:[<получатель>] --- тоже самое, но можно указывать две разных ветки (источником может быть коммит: например, branch_name~3) (при указании несуществующего получателя создаётся новая ветка) (при отсутствии источника удаляется ветка)

    git fetch ; git rebase origin/main ; git push --- обновиться до актуальной версии репозитория для успешного выполнения
        git pull --rebase ; git push
    git fetch ; git merge origin/main ; git push
		git pull ; git push	
    
    git branch -u origin/main [branch_name] --- указать существующей ветке отслеживать origin/main (при выполнении git push в данной ветке изменения будут закачиваться в origin/main)
        git branch -u origin/main --- указать текущей ветке отслеживать origin/main
    git checkout -b [branch_name] origin/main --- создать новую ветку с отслеживанием origin/main
    
#### отслеживание изменений
    git status --- вывести информацию о состоянии файлов в директории и индексе
            untracked (неотслеживаемое) – файл создан, но не добавлен в репозиторий
            modified (изменённое) – изменённые файлы в репозитории
            staged (подготовленное) – отобранные изменённые файлы в репозитории
            committed (закомиченное) – закомиченные файлы в репозитории

    git log --- вывести историю коммитов (по умолчанию - текущей ветки)
        git show [commit] --- вывести информацию о конкретном коммите (достаточно первых символов идентификатора коммита)
    git diff --- вывести перечень изменений в файлах (в состоянии modified); показать разницу между рабочей директорией и индексом
        git diff --staged --- показать разницу между индексом и последним коммитом
        git diff [commit]..[commit] --- показать разницу между двумя коммитами (нужно указать промежуток) 
        git diff [branch1] [branch2]
        
#### другие команды
    git mv [<options>] [source]… [destination] --- переместить файлы, выполнить "git add" для нового файла и "git rm" для старого
    git clean --- удалить ненужные файлы из рабочей директории (результаты сборки проекта или файлы конфликтов слияний)
    
    git tag [tag_name] [commit] --- создать постоянную ссылку на коммит
        git tag [tag_name] --- создать постоянную ссылку на текущий HEAD-коммит
    git describe [ref] --- (ref — что-либо, указывающее на конкретный коммит (например, ветка); по умолчанию - HEAD).
        Вывод команды выглядит так: <tag>_<numCommits>_g<hash>
            tag – ближайший тег в истории изменений,
            numCommits – насколько далеко мы от этого тега,
            hash – хеш коммита, который описывается
